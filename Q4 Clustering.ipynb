```python
#Creating a dictionary of 10 different Cryptocurrencies
Currency_dict = {
    'BTC':'BTC-USD',
    'BSV':'BSV-USD',
    'Ethereum':'ETH-USD',
    'Dogecoin':'DOGE-USD',
    'Theta':'THETA-USD',
    'Binance':'BNB-USD',
    'Solana':'SOL-USD',
    'Terra':'LUNA1-USD',
    'Avalanche':'AVAX-USD',
    'Shiba':'SHIB-USD'}
```

```python
start_time=datetime.datetime(2021,4,1)
end_time=datetime.datetime(2022,4,21)
 
df=data.get_data_yahoo(list(Currency_dict.values()), start=start_time, end=end_time)
df
```

```python
stock_open = np.array(df['Open']).T # stock_open is numpy array of transpose of df['Open']
stock_close = np.array(df['Close']).T # stock_close is numpy array of transpose of df['Close']
```

```python
movements = stock_close - stock_open
sum_of_movement = np.sum(movements,1)
for i in range(len(Currency_dict)):
 print('Currency:{}, Change:{}'.format(df['High'].columns[i],sum_of_movement[i]))
```

```python
plt.figure(figsize = (20,10)) 
plt.subplot(1,2,1) 
plt.title('Currency:BTC',fontsize = 20)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 20)
plt.xlabel('Date',fontsize = 15)
plt.ylabel('Opening price',fontsize = 15)
plt.plot(df['Open']['BTC-USD'])
plt.subplot(1,2,2) 
plt.title('Currency:BSV',fontsize = 20)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 20)
plt.xlabel('Date',fontsize = 15)
plt.ylabel('Opening price',fontsize = 15)
plt.plot(df['Open']['BSV-USD'])
```

```python
# Import the necessary packages
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
# Define a normalizer
normalizer = Normalizer()
# Create Kmeans model
kmeans = KMeans(n_clusters = 3,max_iter = 1000)
# Make a pipeline chaining normalizer and kmeans
pipeline = make_pipeline(normalizer,kmeans)
# Fit pipeline to daily stock movements
pipeline.fit(movements)
labels = pipeline.predict(movements)
```

```python
df1 = pd.DataFrame({'labels':labels,'Currency':list(Currency_dict)}).sort_values(by=['labels'],axis = 0)
df1
```
